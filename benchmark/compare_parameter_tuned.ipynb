```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# === CONFIGURATION ===
BENCHMARK_DIR = os.path.join(os.getcwd(), 'data', 'evaluate')
BENCHMARK_FILES = [
    'evaluation_results_20251229-234204_20251230_114310.csv',
    'evaluation_results_20251230-133406_20251230_135210_tune1.csv',
    'evaluation_results_20251230-143729_20251230_144809_tune2.csv'
]
BENCHMARK_LABELS = [
    'Before Tuning',
    'After Tuning1',
    'After Tuning2'
]
```

```python
# === LOAD DATA ===
dfs = []
for file in BENCHMARK_FILES:
    path = os.path.join(BENCHMARK_DIR, file)
    if os.path.exists(path):
        df = pd.read_csv(path)
        dfs.append(df)
        print(f"Loaded benchmark data from {file}")
    else:
        print(f"Benchmark file {file} not found.")
        dfs.append(None)

# Remove any failed loads
dfs = [df for df in dfs if df is not None]
if len(dfs) < 2:
    print("Need at least two valid benchmark files to compare.")
    raise ValueError("Not enough valid benchmark files.")
```

```python
# === AGGREGATE MEANS AND STDS ===
algo_means = [df.groupby('algorithm').mean() for df in dfs]
algo_stds = [df.groupby('algorithm').std() for df in dfs]

# Find common algorithms
common_algorithms = set(algo_means[0].index)
for means in algo_means[1:]:
    common_algorithms &= set(means.index)
common_algorithms = sorted(list(common_algorithms))

# Get numeric columns
numeric_cols = algo_means[0].select_dtypes(include=[np.number]).columns.tolist()
```

```python
# === PLOTTING: Grouped Metrics in Figures ===

# Exclude 'attempt' and any column containing 'std'

filtered_metrics = [col for col in numeric_cols if 'attempt' not in col.lower() and 'std' not in col.lower()]

metrics_per_fig = 3  # Number of metrics per figure (adjust as needed)
n_metrics = len(filtered_metrics)
n_figs = int(np.ceil(n_metrics / metrics_per_fig))

for fig_idx in range(n_figs):
    fig, axes = plt.subplots(1, metrics_per_fig, figsize=(7 * metrics_per_fig, 6))
    if metrics_per_fig == 1:
        axes = [axes]
    fig.suptitle('Algorithm Performance Comparison', fontsize=16, fontweight='bold')

    for sub_idx in range(metrics_per_fig):
        metric_idx = fig_idx * metrics_per_fig + sub_idx
        if metric_idx >= n_metrics:
            axes[sub_idx].axis('off')
            continue
        metric = filtered_metrics[metric_idx]
        ax = axes[sub_idx]
        x = np.arange(len(common_algorithms))
        width = 0.8 / len(dfs)
        colors = plt.cm.get_cmap('tab10', len(dfs))

        for i, means in enumerate(algo_means):
            values = means.loc[common_algorithms, metric].values
            ax.bar(x + (i - (len(dfs)-1)/2) * width, values, width, 
                   label=BENCHMARK_LABELS[i] if i < len(BENCHMARK_LABELS) else f"Run {i+1}",
                   alpha=0.85, color=colors(i))

        ax.set_xlabel('Algorithm', fontweight='bold')
        ax.set_ylabel(metric, fontweight='bold')
        ax.set_title(metric)
        ax.set_xticks(x)
        ax.set_xticklabels(common_algorithms, rotation=45, ha='right')
        ax.legend()
        ax.grid(axis='y', alpha=0.3)

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

```python
# === IMPROVEMENT SUMMARY ===
print("\n=== Performance Improvement Summary ===")
for metric in numeric_cols:
    print(f"\n{metric}:")
    for i in range(1, len(algo_means)):
        improvement = ((algo_means[i].loc[common_algorithms, metric] - algo_means[0].loc[common_algorithms, metric]) /
                       algo_means[0].loc[common_algorithms, metric] * 100).round(2)
        for algo in improvement.index:
            print(f"  {algo} ({BENCHMARK_LABELS[0]}  {BENCHMARK_LABELS[i] if i < len(BENCHMARK_LABELS) else f'Run {i+1}'}): {improvement[algo]:+.2f}%")
```
